
## Simple Epidemic
Consider a closed population population of N individuals. <br>
There are two states:<br>
- Susceptible
- Infected <br>


Initially $ I_0 $ are infected ,
$N − I_0 $ are therefore susceptible .
We assume the population is well mixed.
The probability that a susceptible and infectious individual meet is proportional to
their abundances, with effective transmission rate $\beta$
In simple model ,we consider the population (N) is closed and mixed.<br>


We introduce, $ s = \frac{S}{N} $and $ i = \frac{I}{N} $<br>
$$\frac{di}{dt} = \beta i(1 − i)$$


To calculate the number infected at time t, i(t), integrate this equation from time
zero to time t, yielding:

$$i(t) = \frac{1}{1 + \frac{1−i_0}{i_ 0} e^{−\beta t} }$$ 

This equation yields what is known as the epidemic curve.

```python
import numpy as np
i_0=0.0001# initial fraction of infected
beta=0.11# effective transmission rate
def i(t,a0,b):
    return 1/(1+ (np.exp(-b*t)*(1-a0)/a0))
import matplotlib.pyplot as plt
t=np.linspace(0,150,100)
plt.plot(t,i(t,i_0,beta),"b--")
plt.xlabel("Days")
plt.ylabel("Cumulative Fraction Infected")
plt.show()
```

# Interpreting the Epidemic Curve
The above  figure plots the cumulative prevalence of the infection.We might also want to know about the shape of the incidence of infection, that is,the number of new cases per unit time.

```python
plt.plot(t,i(t,i_0,beta)-i(t-1,i_0,beta),"r--")
plt.xlabel("Days")
plt.ylabel("Incident Fraction Infected")
plt.title("Epidemic Curve")
plt.show()
```

# More Interpretations of the Epidemic Curve
This is the classic epidemic curve.
The epidemic curve is “bell-shaped”, but not completely symmetric.
There is a greater force of infection early on.
Note that in the limit $ t \to \infty $, everyone in the population becomes infected.
